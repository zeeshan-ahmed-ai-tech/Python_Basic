```python
#for loop

for i in range(1,6):
  print(i)
```

```python
#while loop

i = 1
while i <=5:
  print(i)
  i+=1
```

```python
#continue
for i in range(1,6):
  if i == 3:
   continue
  print(i)
```

```python
#break
for i in range(1,6):
  if i == 4:
    break
  print(i)
```
